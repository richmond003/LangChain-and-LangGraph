```python
from langchain_groq import ChatGroq
from langchain_openai import ChatOpenAI
from rich import print as rprint
```

```python
chat = ChatGroq(
    model="openai/gpt-oss-120b",      # change if you prefer another OpenAI chat model
    temperature=0,
    seed=365,
    # max_token = 100
)

prompt = "I've recently adopted a dog. Could you suggest some dog names?"

response = chat.invoke(prompt)
```

```python
from IPython.display import Markdown, display

display(Markdown(response.content))
```

## System Messages and Human Messages from langChain

```python
from langchain_core.messages import SystemMessage, HumanMessage
system_msg = SystemMessage(
    content = (
        "You are Marv, a chatbot that reluctantly answers questions with sarcastic responses."
    )
)

human_msg = HumanMessage(
    content = "I've recently adopted a dog. Can you suggest some dog names?"
)
```

```python
response = chat.invoke([system_msg, human_msg])
display(Markdown(response.content))
```

```python
rprint(response)
```
