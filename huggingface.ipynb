# Pre-trained Tokenizers

```python
from transformers import AutoTokenizer
model = "bert-base-uncased"
sentence = "working with pre-trained tokenizers from huggingface"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
from rich.pretty import pprint
input_ids = tokenizer(sentence)
pprint(input_ids)
```

```python
tokens = tokenizer.tokenize(sentence)
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(token_ids)
```

```python
print(tokenizer.decode(input_ids["input_ids"]))
print(tokenizer.decode(token_ids))
```

```python
xlnet_model = "xlnet-base-cased"
```

```python
xlnet_tokenizer = AutoTokenizer.from_pretrained(xlnet_model)
```

```python
xlnet_input_ids = xlnet_tokenizer(sentence)
pprint(xlnet_input_ids)
```

```python
xlnet_tokens = xlnet_tokenizer.tokenize(sentence)
pprint(xlnet_tokens)
```

```python
xlnet_token_ids = xlnet_tokenizer.convert_tokens_to_ids(xlnet_tokens)
pprint(xlnet_token_ids)
```

```python
pprint(xlnet_tokenizer.decode(xlnet_token_ids))
pprint(xlnet_tokenizer.decode(xlnet_input_ids["input_ids"]))
```

# HuggingFace and Pytorch

```python
pos_sentence = "I am doing welling in most of my classes."
neg_sentence = "I am feeling under the weather today."
```

```python
import torch
```

```python
bert_model = "distilbert-base-uncased-finetuned-sst-2-english"
```

```python
tokenizer = AutoTokenizer.from_pretrained(bert_model)
```

```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained(bert_model)
```

```python
# get input_ids as tensors
def get_tensors_input_ids(sentence):
    input_ids_pt = tokenizer(sentence, return_tensors="pt")
    return input_ids_pt
```

```python
pos_input_ids_tensor = get_tensors_input_ids(pos_sentence)
neg_input_ids_tensor = get_tensors_input_ids(neg_sentence)
```

```python
def get_seqence_sentiment(input_ids_pt):
    with torch.no_grad():
        logits = model(**input_ids_pt).logits

    predicted_class_id = logits.argmax().item()
    return model.config.id2label[predicted_class_id]
```

```python
pprint(f"{pos_sentence} : {get_seqence_sentiment(pos_input_ids_tensor)}")
pprint(f"{neg_sentence} : {get_seqence_sentiment(neg_input_ids_tensor)}")
```
