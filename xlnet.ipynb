# Preprocessing Data

```python
data_train = pd.read_csv(r"datasets\emotion-labels-train.csv")
data_test = pd.read_csv(r"datasets\emotion-labels-test.csv")
data_val = pd.read_csv(r"datasets\emotion-labels-val.csv")
```

```python
# concat test, val and train set for preprocessing process
data = pd.concat([data_train, data_test, data_val], ignore_index=True)
```

```python
# removing emoji and punctuation from datasets
from emoji import replace_emoji
data["text_clean"] = data["text"].apply (lambda x: clean(x, punct=True))
data["text_clean"] = data["text_clean"].apply(lambda text: replace_emoji(text, replace=" "))
```

```python
# checking how may labels are associated with the dataset and each counts
data["label"].value_counts().plot(kind="bar")
```

```python
# balancing dataset
gd_labels = data.groupby('label')
data = pd.DataFrame(gd_labels.apply(lambda x: x.sample(gd_labels.size().min()).reset_index(drop=True)))
```

```python
data["label_int"] = LabelEncoder().fit_transform(data["label"])
```
