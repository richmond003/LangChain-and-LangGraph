```python
from langchain_groq import ChatGroq
from langchain_openai import ChatOpenAI
from rich import print as rprint
```

```python
chat = ChatOpenAI(
    model="gpt-5.4-mini-2026-03-17",      # change if you prefer another OpenAI chat model
    temperature=0,
    seed= 365,
    max_tokens = 100
)

prompt = "I've recently adopted a dog. Could you suggest some dog names?"

response = chat.invoke(prompt)
```

## System Messages and Human Messages from langChain

```python
from langchain_core.messages import SystemMessage, HumanMessage
system_msg = SystemMessage(
    content = (
        "You are Marv, a chatbot that reluctantly answers questions with sarcastic responses."
    )
)

human_msg = HumanMessage(
    content = "I've recently adopted a dog. Can you suggest some dog names?"
)
```

```python
response = chat.invoke([system_msg, human_msg])
display(Markdown(response.content))
```

```python
from langchain_core.prompts import (
    SystemMessagePromptTemplate, 
    HumanMessagePromptTemplate, 
    ChatPromptTemplate)

SYSTEM_PROMPT_TEMPLATE = '{description}'

HUMAN_PROMPT_TEMPLATE = """ 
I've recently adopted a {pet}. Could you
suggest some {pet} names?
 """
```

```python
system_msg_temaplate = SystemMessagePromptTemplate.from_template(template=SYSTEM_PROMPT_TEMPLATE)
human_msg_template = HumanMessagePromptTemplate.from_template(template=HUMAN_PROMPT_TEMPLATE)
```

```python
rprint(system_msg_temaplate)
```

```python
chat_template = ChatPromptTemplate.from_messages([system_msg_temaplate, human_msg_template])
```

```python
rprint(chat_template)
```

```python
sys_prompt = """ You are a chatbot that answers questions in the way Master Yoda from Star Wars would. """
chat_value = chat_template.invoke({"description": sys_prompt, "pet": "Dog"})
```

```python
rprint(chat_value)
```

```python
response = chat.invoke(chat_value)
```

```python
rprint(response)
```

```python
display(Markdown(response.content))
```

# Few shot prompting

```python
from langchain_core.prompts import FewShotChatMessagePromptTemplate, AIMessagePromptTemplate
from langchain.chains import LLMChain
```

```python
TEMPLATE_H = '''I've recently adopted a {pet}. 
Could you suggest some {pet} names?'''
TEMPLATE_AI = '''{response}'''

message_template_h = HumanMessagePromptTemplate.from_template(template = TEMPLATE_H)
message_template_ai = AIMessagePromptTemplate.from_template(template = TEMPLATE_AI)
```

```python
rprint(message_template_ai)
rprint(message_template_h)
```

```python
example_template = ChatPromptTemplate.from_messages([message_template_h, message_template_ai])
```

```python
rprint(example_template)
```

```python
examples = [{'pet':'dog', 
             'response':'''Oh, absolutely. Because nothing screams "I'm a responsible pet owner" 
like asking a chatbot to name your new furball. How about "Bark Twain" (if it's a literary hound)? '''}, 
            
            {'pet':'cat', 
             'response':'''Oh, absolutely. Because nothing screams "I'm a unique and creative individual" 
             like asking a chatbot to name your cat. How about "Furry McFurFace", "Sir Meowsalot", or "Catastrophe"? '''}, 
            
            {'pet':'fish', 
             'response':
             '''Oh, absolutely. Because nothing screams "I'm a fun and quirky pet owner" 
             like asking a chatbot to name your fish. How about "Fin Diesel", "Gill Gates", or "Bubbles"?'''}]
```

```python
few_shot_prompt = FewShotChatMessagePromptTemplate(
    examples= examples,
    example_prompt=example_template,
    input_variables=['pet']
)
```

```python
rprint(few_shot_prompt)
```

```python
chat_template = ChatPromptTemplate.from_messages([few_shot_prompt, message_template_h])
```

```python
rprint(chat_template)
```
