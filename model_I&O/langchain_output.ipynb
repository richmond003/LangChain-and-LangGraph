# Output Parser

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from rich import print as rprint
```

```python
OPENAI_MODELS = {
    "GPT5.4_mini" : "gpt-5.4-mini-2026-03-17",
    "GPT5.5" : "gpt-5.5-2026-04-23",
}
def create_chat(model, temperature = 0, seed = 365, max_tokens = 500):
    return ChatOpenAI(
        model=model,
        temperature= temperature,
        seed= seed,
        max_tokens=max_tokens
    )
```

## Str Output Parser 

```python
from langchain_core.output_parsers import StrOutputParser
chat = create_chat(model=OPENAI_MODELS["GPT5.4_mini"])
human_msg = HumanMessage(
    content= "Can you give me an interesting fact most people don't know?"
)
```

```python
response = chat.invoke([human_msg])
rprint(response)
```

```python
str_output_parser = StrOutputParser()
res = str_output_parser.invoke(response)
```

```python
print(res)
```

## Comma-Separated List Output Paser

```python
from langchain_core.output_parsers import CommaSeparatedListOutputParser

chat = create_chat(model=OPENAI_MODELS["GPT5.4_mini"])

human_msg = HumanMessage(
    content= f""" 
    Can you suggest some pet name for my dog?

    {CommaSeparatedListOutputParser().get_format_instructions()}
    """
)

response = chat.invoke([human_msg])
```

```python
comman_separated_parser = CommaSeparatedListOutputParser()
parsed = comman_separated_parser.invoke(response)
```

```python
rprint(parsed)
```

## Datetime Parser

```python
from langchain_core.output_parsers import DatetimeOutputParser
chat = create_chat(model=OPENAI_MODELS["GPT5.4_mini"])

human_msg = f""" 
    When was Game of Thrones released

    {DatetimeOutputParser().get_format_instructions()}
"""

response = chat.invoke([human_msg])
```

```python
datetime_parser = DatetimeOutputParser()

parsed = datetime_parser.invoke(response)
print(parsed)
```
